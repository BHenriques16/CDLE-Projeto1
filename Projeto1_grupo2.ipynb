```python
from elasticsearch import Elasticsearch

# Ligação com cabeçalhos de compatibilidade explícitos
es = Elasticsearch(
    "http://localhost:9200",
    headers={"Accept": "application/vnd.elasticsearch+json; compatible-with=8"}
)

# Testar a pesquisa
try:
    res = es.search(index="noticias_tecnologia", query={"match_all": {}})
    print(f"Sucesso! Total de notícias: {res['hits']['total']['value']}")
except Exception as e:
    print(f"Ainda temos um erro: {e}")
```

```python
# Pesquisar por uma palavra específica no título
termo = "Oppo" 

query = {
    "match": {
        "titulo": termo
    }
}

res = es.search(index="noticias_tecnologia", query=query)

print(f"Resultados para '{termo}':")
for hit in res['hits']['hits']:
    print(f"-> {hit['_source']['titulo']} (Fonte: {hit['_source']['fonte']})")
```
