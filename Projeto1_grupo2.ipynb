## Tarefa 1.2

Carregamento do ficheiro JSON em memória e iteração pelos 5 primeiros registos

```python
import json
import os

articles_path = 'data/articles.json'

# Load Json data in memory
if os.path.exists(articles_path):
    with open(articles_path, 'r', encoding='utf-8') as f:
        articles = json.load(f)

    print(f"File '{articles_path}' loaded successfully.")
    print(f"Total articles in memory: {len(articles)}")

    # Iterate through the first 5 articles and print their titles
    for i, article in enumerate(articles[:5]):
        print(f"Article {i + 1}:")

        # json.dumps() to print the article in a better format
        print(json.dumps(article, indent=4, ensure_ascii=False))

else:
    print(f"File '{articles_path}' not found. Please check the path and try again.")
```

## Tarefa 2.1

Indexação dos artigos coletados no Redis

```python
import redis
import json

# Connect to Redis
r = redis.Redis(host='localhost', port=6379, decode_responses=True)

try:
    r.ping()
    print("Connected to Redis successfully.")
except redis.ConnectionError:
    print("Failed to connect to Redis. Please ensure Redis is running and try again.")

# Store articles in Redis
counter = 0

for article in articles:
    counter += 1
    # Create a unique key for each article
    key = f"article:{counter}"

    # Store the article as a JSON string in Redis
    r.set(key, json.dumps(article, ensure_ascii=False))

print(f"Stored {counter} articles in Redis.")
```

## Tarefa 2.2

Recuperação de informação com recurso a queries

```python
# Query 1: count the total number of articles stored in Redis
article_keys = r.keys('article:*')
print(f"Query 1: Total articles stored in Redis: {len(article_keys)}\n")

# Query 2: Retrieve a specific article by its key (e.g., article:3)
article_key = 'article:120'
article_data = r.get(article_key)

if article_data:
    article_retrieved = json.loads(article_data)
    print(f"Query 2: Article retrieved for key '{article_key}':")
    print(json.dumps(article_retrieved, indent=4, ensure_ascii=False))
# Query 3: Search for articles that contain a specific keyword in their title (e.g., "Tecnologia, Inteligencia Artificial")
keyword = "Inteligência Artificial"
matching_articles = []

for key in article_keys:
    doc  = json.loads(r.get(key))
    title = doc.get('titulo', '')

    # Check if the keyword is in lowercase
    if keyword.lower() in title.lower():
        matching_articles.append(doc)

print(f"Query 3: Articles containing the keyword '{keyword}' in their title:")
for i in matching_articles[:3]:  # Print the first 3 matching articles
    print(json.dumps(i, indent=4, ensure_ascii=False))
```
